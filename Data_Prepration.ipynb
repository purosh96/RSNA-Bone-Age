```python
import pandas as pd
```

```python
df = pd.read_csv("boneage-dataset-labels.csv")
df.shape
df = df.set_index('id')
```

```python
#Bone age is give in Months hence 
df["age_float"] = df.boneage/12
df["age_int"] = df.boneage/12
```

```python
df1 = df[df.age_float>20]
# The Age Group Given is 0-19 Year old
```

```python
df.age_int = df.age_int.astype("int32")
```

```python
df = df.sort_values(by=['age_int'])
```

```python
#Given data/img value count as per age
df.age_int.value_counts() 
```
